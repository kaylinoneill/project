```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/Users/sadie/Desktop/stats/CountryData (1).csv')
dataset
```

```python
dataset.dropna(inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
region = dataset["Region"].values.reshape(-1, 1)
enc.fit(region)
region_OneHot = enc.transform(region)
region_one_hot = region_OneHot.toarray()
region_labels = enc.get_feature_names_out()
dataset[region_labels] = region_one_hot
dataset = dataset.drop(columns=["Region","Country"])
```

```python
X = dataset.loc[:, dataset.columns != "GDP"]
y = dataset["GDP"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
guess = reg.predict(X_test)
```

```python
mean_squared_error(y_test,guess)
```

```python
reg.score(X_test, y_test)
```
